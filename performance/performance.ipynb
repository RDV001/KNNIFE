```python
import sys
import os
sys.path.append('..')

import numpy as np

from MyEnsemble import MyIF, MyEIF, MyNNIF, MyDIF, fit_stats, fit_scores

import my_datasets

import pandas as pd
```

Performance function

```python
def fit_performance(
    model_class,
    random_state, 
    iter
):

    folder = model_class.__name__
    os.makedirs(folder, exist_ok=True) 
    
    for dataset_name in my_datasets.datasets_names:

        data_dict = my_datasets.load(dataset_name)
        X = data_dict['X']
        y = data_dict['y']

        model_stats = pd.DataFrame(columns=fit_scores)

        for i in range(iter):

            model = model_class(contamination=np.mean(y), random_state=random_state+i)
            model.fit(X)
            stats_buffer = fit_stats(model, y)
            model_stats = pd.concat([model_stats, stats_buffer], ignore_index=True)

        model_stats_mean = model_stats.mean()
        model_stats_sem = model_stats.sem()
        model_stats_std = model_stats.std(ddof=1)

        if not os.path.isdir(f'{folder}\{dataset_name}'):
            os.makedirs(f'{folder}\{dataset_name}')
        model_stats.to_csv(f'{folder}\{dataset_name}\{dataset_name}.csv')
        model_stats_mean.to_csv(f'{folder}\{dataset_name}\{dataset_name}_mean.csv')
        model_stats_sem.to_csv(f'{folder}\{dataset_name}\{dataset_name}_sem.csv')
        model_stats_std.to_csv(f'{folder}\{dataset_name}\{dataset_name}_std.csv')
```

Function call

```python
fit_performance(model_class=MyIF, random_state=0, iter=10)
```
