```python
import numpy as np

from MyEnsemble import MyIF, MyEIF, MyNNIF, MyDIF, fit_stats

from MyInterpreter import MyInterpreter

import my_datasets
```

Dataset

```python
data_dict = my_datasets.load(dataset_name="secom", )

X = data_dict['X']
y = data_dict['y']
name = data_dict['name']

np.sum(y)
```

Isolation Forest

```python
IF = MyIF(
    contamination = np.mean(y),
    random_state = 0
)

IF.fit(X)

fit_stats(IF, y)
```

Extended Isolation Forest

```python
EIF = MyEIF(
    contamination = np.mean(y),
    random_state = 0
)

EIF.fit(X)

fit_stats(EIF, y)
```

NNIF

```python
NNIF = MyNNIF(
    contamination = np.mean(y),
    random_state = 0
)

NNIF.fit(X)

fit_stats(NNIF, y)
```

Deep Isolation Forest

```python
DIF = MyDIF(
    contamination = np.mean(y),
    random_state = 0
)

DIF.fit(X)

fit_stats(DIF, y)
```

Interpretation

```python
interpreter = MyInterpreter(method='kNN', ensemble=IF)
```
