```python
import pandas as pd

normal_data = pd.read_csv('my_datasets/TEP_FaultFree_Training_subsample_70_3.csv')
fault_data = pd.read_csv('my_datasets/TEP_Faulty_Training_subsample_70_3_removedfirst20.csv')

faulty_data, data = [], []
for i in range(20): 
    faulty_data.append(fault_data[fault_data['faultNumber'] == i+1].reset_index(drop=True))
    data.append(pd.concat([normal_data, faulty_data[i]]).reset_index(drop=True))

features = data[0].columns[3:] 

for i in range(20):
    data[i][features] = data[i][features].sample(frac=1).reset_index(drop=True)
    data[i][features].to_csv(f"TEP_{i+1}.csv")
```
