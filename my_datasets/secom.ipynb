```python
import numpy as np
import pandas as pd
```

```python
with open("secom.data", "r") as infile:
    rows = []
    for i in infile:
        row = [x.lower() for x in i.split(" ")]
        row[-1] = row[-1][0:-1]
        row = [float(x) for x in row]
        rows.append(row)

with open("secom_labels.data", "r") as infile:
    labels = []
    for i in infile:
        label = int(int([x.lower() for x in i.split(" ")][0]) > 0)
        labels.append(label)

data = np.vstack(rows)
labels = np.vstack(labels)

data = np.hstack((data, labels))
data = np.nan_to_num(data)

data_abs = np.abs(data)
data_mean = np.mean(data_abs, axis=0)
data_diff = data - data_mean
data_sum = np.sum(data_diff, axis=0)

data = data[:, data_sum != 0]

pd.DataFrame(data).sample(frac=1).reset_index(drop=True).to_csv("secom.csv")
```
